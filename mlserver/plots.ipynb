```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set_palette(sns.color_palette("bright"))
```

```python
many_models = np.loadtxt("many_models.csv", delimiter=",", skiprows=1)
single_model = np.loadtxt("single_model.csv", delimiter=",", skiprows=1)
hops = [1, 2, 5, 10, 20, 40, 60, 80]

many_models = many_models / 1000000.0
single_model = single_model / 1000000.0
```

```python
def plot_median_quantiles(x, y, color, label):
    plt.fill_between(x, np.quantile(y, q=0.1, axis=1), np.quantile(y, q=0.9, axis=1), facecolor=color, alpha=0.15)
    plt.plot(x, np.median(y, axis=1), c=color, label=label)
```

```python
x = range(len(hops))
plot_median_quantiles(x, many_models, 'r', 'Many hops')
plot_median_quantiles(x, single_model, 'b', 'No hops')
plt.legend()
```
